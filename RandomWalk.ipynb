```python
def RanWal1D(n_steps):
    prob=0.5
    pos=np.zeros(n_steps)
    for i in range(1,n_steps):
        x=np.random.rand()
        if x<prob:
            pos[i]=pos[i-1]-1
        else:
            pos[i]=pos[i-1]+1
    return pos
```

```python
positions=RanWal1D(1000);
steps=np.arange(0,10000,1)
```

```python
plt.plot(steps,positions)
```

```python
def RWmul(n_steps, n_sim):
    prob=0.5
    pos=np.zeros((n_sim,n_steps))
    for j in range(0,n_sim):
        for i in range(1,n_steps):
            x=np.random.rand()
            if x<prob:
                pos[j,i]=pos[j,i-1]-1
            else:
                 pos[j,i]=pos[j,i-1]+1
    return pos
        
```

```python
simu=RWmul(10000,1000)
```

```python
plt.plot(steps,simu[999,:])
```

```python
for i in range(1000):
    plt.plot(steps, simu[i, :], linewidth=1, alpha=0.5) 
plt.xlabel('number of steps')
plt.ylabel('position')
```
