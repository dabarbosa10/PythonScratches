```python
#define a function that takes the number of trials and return ne number of hits
def hits(N):
    Nhits=0
    for i in range(N):
        x=2*np.random.random_sample()-1
        y=2*np.random.random_sample()-1
        if (x**2+y**2<1):
            Nhits=Nhits+1
            plt.plot(x,y,'g.')
    pic=4*(Nhits)/N
    return Nhits, pic
```

```python
hits(4000)
```

```python
for i in range(5):
    print (hits(4000))
```

In theory the result of $\pi$ improves and becomes exact in the limit of an infinite number of trial. Due to the Bernoullis weak law of large numbers. 

Now, let's change the algorithm to a Markov Chaing sampling. Any move that take us outside the square is rejected. In other words, we don't move and count the configuration a second time.

```python
def markovpi(N):
    Nhits=0;
    x=1;
    y=1;
    for i in range(N):
        deltax=np.random.random_sample()-0.5
        deltay=np.random.random_sample()-0.5
        if(np.abs(x+deltax)<1 and np.abs(y+deltay)<1):
            x=x+deltax
            y=y+deltay
        if(x**2+y**2<1):
            Nhits=Nhits+1
            plt.plot(x,y,'.')
    pic=4*(Nhits)/N
    return Nhits,pic
```

```python
for i in range(5):
    print(markovpi(4000))
```

Here, we have to approaches to sampling a probability distribution $\pi(x)$ on a discrete or continuous space: Direct Sampling and Markov-chain sampling.
